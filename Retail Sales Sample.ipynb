```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data_sample.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['PRICEEACH'].plot(kind='hist', bins=10, title='Histogram of item price ordered', xlabel='Price');
plt.savefig('Histogram of item price ordered.jpg')
```

```python
ax = df.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().plot(kind='bar', title='Quantity Ordered for each Product Line', xlabel='Product Line', ylabel='Quantity Ordered',
                                                             rot=0, figsize=(11,7))
ax.bar_label(ax.containers[0]);
plt.savefig('Quantity ordered for each product line.jpg')
```

```python
df.groupby('DEALSIZE')['QUANTITYORDERED'].sum().plot(kind='pie', autopct='%.1f%%', title='Sizes of Product Ordered', ylabel='', legend=True, labels=None, figsize=(6,6))
plt.savefig('Sizes of product ordered.jpg')
```

```python
df['STATUS'].value_counts()
```

```python
ab = df['STATUS'].value_counts().plot(kind='bar', title='Status of Products Ordered', xlabel='Staus', ylabel='Count', rot=0)
ab.bar_label(ab.containers[0]);
plt.savefig('Status of Products Ordered.jpg')
```

```python
ab = df['PRODUCTLINE'].value_counts().plot(kind='bar', title='Product Line Ordered', xlabel='Product Line', ylabel='Count', rot=0, figsize=(11,7))
ab.bar_label(ab.containers[0]);
plt.savefig('Product line ordered.jpg')
```

```python
ab = df['DEALSIZE'].value_counts().plot(kind='pie', autopct='%.1f%%', title='Sizes Ordered', ylabel='', legend=True, labels=None, figsize=(6,6))
plt.savefig('Sizes Ordered.jpg')
```
